```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
data
```

```python
dataset = pd.DataFrame(data = data['data'], columns = data['feature_names'])
dataset
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
# X is the features dataset and y is the target dataset
X = dataset.copy()
y = pd.DataFrame(data = data['target'], columns = ['target'])
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(ccp_alpha = 0.01)
clf = clf.fit(X_train, y_train)
```

```python
clf.get_params()
```

```python
X_test
```

```python
predictions = clf.predict(X_test)
predictions
```

```python
X_test1 = X_test.reset_index(drop=True)
X_test1
```

```python
clf.predict_proba(X_test1)
```

```python
predict_data = pd.concat([X_test1, pd.DataFrame(data = predictions, columns = ['target'])], axis = 1)
```

```python
predict_data
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions, labels=[0,1])
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions, target_names = ['malignant', 'benign']))
```

```python
feature_names = X.columns
feature_names
```

```python
feature_importance = pd.DataFrame(data = clf.feature_importances_, index = feature_names, columns = ['score'])
feature_importance = feature_importance.sort_values(by = 'score', ascending = False)
```

```python
feature_importance.head(10).plot(kind='bar')
```

```python
from sklearn import tree
from matplotlib import pyplot as plt

fig = plt.figure(figsize = (25,20))
fig = tree.plot_tree(clf, 
               feature_names = feature_names, 
               class_names = {0 : 'Malignant', 1 : 'Benign'}, filled = True, fontsize = 12)
```
